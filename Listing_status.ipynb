```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("airbnb_listing_from_2018.csv")
df.head()
```

```python
df["year_month"] = df["year"].astype(str) +"_"+ df["month"].astype(str)
```

```python
pivot_df = df[["id","year_month","file_name"]].pivot(index="id",columns="year_month",values="file_name")
pivot_df.head()
```

```python
def isNaN(x):
    if str(x) == 'nan':
        return 0
    return 1
```

```python
for col in pivot_df:
    pivot_df[col] = pivot_df.apply(lambda x : isNaN(x[col]),axis=1)
pivot_df.head()
```

```python
ordered_col_list = []
years = df["year"].unique()
years.sort()
months = [i for i in range(1,13)]
for i in years:
    for j in months:
        if str(i) + "_" + str(j) in list(pivot_df.columns):
            ordered_col_list.append(str(i) + "_" + str(j))
ordered_col_list
```

```python
pivot_df = pivot_df[ordered_col_list]
pivot_df.head()
```

```python
def isNew(x,y):
    if y == 1 and x == 0:
        return "New"
    if y == 0 and x == 0:
        return "NA"
    if y == 1 and x == 1:
        return "Old"
    if y == 0 and x == 1:
        return "Dropped"
```

```python
pivot_df_status = pivot_df.copy()
```

```python
for i in range(0,len(ordered_col_list)):
    if i == 0:
        continue
    pivot_df_status[ordered_col_list[i]] = pivot_df.apply(lambda x: isNew(x[ordered_col_list[i-1]],x[ordered_col_list[i]]),axis=1)
pivot_df_status.head()
```

```python
def isExist(x):
    if x == 0:
        return "NA"
    return "New"
```

```python
pivot_df_status['2018_3'] = pivot_df.apply(lambda x : isExist(x['2018_3']),axis=1)
```

```python
pivot_df_status.head()
```

```python
pivot_df_status_flat = pd.DataFrame(columns=["id","year","month","status"],index=range(len(pivot_df.index) * len(pivot_df.columns)))
print(pivot_df_status_flat.shape)
c = 0
for i in pivot_df_status.index:
    for j in pivot_df_status.columns:
        pivot_df_status_flat.iloc[c,0] = i
        yms = j.split("_")
        pivot_df_status_flat.iloc[c,1] = yms[0]
        pivot_df_status_flat.iloc[c,2] = yms[1]
        pivot_df_status_flat.iloc[c,3] = pivot_df_status.at[i,j]
        c = c + 1
        print(c,end="\r")
pivot_df_status_flat.head()
```

```python
pivot_df_status_flat.to_csv("listings_status_flat.csv",index=False)
```
