```python
import pandas as pd
import io
import csv
import os
import datetime as dt
import numpy as np
```

```python
cols_list = ['id',
'host_id',
'host_response_time',
'host_response_rate',
'host_acceptance_rate',
'host_listings_count',
'host_total_listings_count',
'host_verifications',
'host_has_profile_pic',
'host_identity_verified',
'neighbourhood',
'neighbourhood_cleansed',
'zipcode',
'property_type',
'room_type',
'accommodates',
'bathrooms',
'bedrooms',
'beds',
'amenities',
'price',
'cleaning_fee',
'minimum_nights',
'has_availability',
'number_of_reviews',
'review_scores_rating',
'review_scores_accuracy',
'review_scores_cleanliness',
'review_scores_checkin',
'review_scores_communication',
'review_scores_location',
'review_scores_value',
'cancellation_policy',
'reviews_per_month',
'calculated_host_listings_count_entire_homes',
'calculated_host_listings_count_private_rooms',
'calculated_host_listings_count_shared_rooms',]
```

```python
dir_path = "./Drive/Data Files [Detailed] "
all_file_path = []
all_file_names = []
all_file_year = []
all_file_month = []
for i in range(1,6):
    paths = [x for x in os.walk(dir_path + str(i))]
    root, dir, files = paths[0]
    for j in files:
        if j.__contains__("_listings"):
            all_file_path.append(root + "/" + j)
            all_file_names.append(j)
            all_file_year.append(int(j.split("_")[2]))
            all_file_month.append(int(j.split("_")[3]))
```

```python
path_df = pd.DataFrame(all_file_path,columns=["path"])
names_df = pd.DataFrame(all_file_names,columns=["name"])
year_df = pd.DataFrame(all_file_year,columns=["year"])
month_df = pd.DataFrame(all_file_month,columns=["month"])
file_paths_df = pd.concat([path_df,names_df,year_df,month_df],axis=1)
file_paths_df.head()
```

```python
file_paths_df = file_paths_df.sort_values(by=["year","month"])
file_paths_df = file_paths_df.loc[file_paths_df["year"]>2017]
file_paths_df.reset_index(drop=True,inplace=True)
file_paths_df.head()
```

```python
file_paths_df.shape
```

```python
num_files = file_paths_df.shape[0]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.DataFrame()
for i in range(0,num_files):
    loop_start_time  = dt.datetime.now()
    temp_df = pd.read_csv(file_paths_df.iloc[i,0])
    print(i,"Reading Finished",temp_df.shape,"duration: ", dt.datetime.now() - loop_start_time)
    all_cols = list(temp_df.columns)
    use_cols = list(set(all_cols).intersection(set(cols_list)))
    temp_df = temp_df[use_cols]
    pending_cols = set(cols_list).difference(set(use_cols))
    for j in pending_cols:
        temp_df[j] = np.nan
    temp_df = temp_df[cols_list]
    temp_df['year'] = file_paths_df.iloc[i,2]
    temp_df['month'] = file_paths_df.iloc[i,3]
    temp_df['file_name'] = file_paths_df.iloc[i,1]
    df = pd.concat([df,temp_df],axis=0)
    df.drop_duplicates(inplace=True)
    print(i,"Cleaning Finished",temp_df.shape,"duration: ", dt.datetime.now() - loop_start_time)
```

```python
def replace_comma(x):
    if type(x) == str:
        return x.replace(",", ";" )
    else:
        return x
```

```python
# this is neccessary bcs the csv file written by pandas has default delimiter(,) 
# but if any column contains that then the data gets corrupted. 
# So this ensures that all (,) is replaced by (;)
for col in df:
    print(col)
    df[col] = df.apply(lambda x : replace_comma(x[col]),axis=1)
```
